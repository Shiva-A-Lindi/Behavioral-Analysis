## IgorProIO Demo

```python
%matplotlib inline
import matplotlib.pyplot as plt
import neo
```

### IBW support

```python
# Downloaded from Human Brain Project Collaboratory
# Digital Reconstruction of Neocortical Microcircuitry (nmc-portal)
# http://microcircuits.epfl.ch/#/animal/8ecde7d1-b2d2-11e4-b949-6003088da632
x = neo.io.IgorIO(filename='nmc-portal/grouped_ephys/B95/B95_Ch0_IDRest_107.ibw')
signal = x.read_analogsignal()
plt.plot(signal.times,signal)
plt.xlabel(signal.sampling_period.dimensionality)
plt.ylabel(signal.dimensionality);
```

### PXP support

```python
# Same as above, but the .ibw file was loaded into Igor and saved as part of a .pxp file.  
x = neo.io.IgorIO(filename='nmc-portal/grouped_ephys/B95/B95_Ch0_IDRest_107.pxp')
signal = x.read_analogsignal(path='root:AA_IDRest_Ch0_107')
plt.plot(signal.times,signal)
plt.xlabel(signal.sampling_period.dimensionality)
plt.ylabel(signal.dimensionality);
```
